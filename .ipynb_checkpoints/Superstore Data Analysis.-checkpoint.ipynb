```python
sps_data_product_Cat = sps_data.groupby(['Category']).sum()
print(sps_data_product_Cat)
```

```python
a = sps_data_product_Cat.loc[:, sps_data_product_Cat.columns.intersection(['Sales'])]
print(a)
```

```python
sps_data_product_count = sps_data.groupby(['Category']).count()
```

```python
b= sps_data_product_count.loc[:, sps_data_product_count.columns.intersection(['Product Name'])]
print(b)
```

```python
d = a.merge(b, on='Category')
print(d)
```

```python
c =sps_data[['Category','Segment']]
print(c)
```

```python
c.Segment
```

```python
uniCat = c.Category.unique()
uniHeader = c.Segment.unique()
colVal = []
for seg in uniHeader:
    for cat in uniCat:
        dd = cat+seg
        colVal.append(c.value_counts()[cat, seg])
    d[seg] = colVal
    colVal = []
```

```python
d
```

```python
 sps_data_regionVsCat = sps_data[['Category','Region', 'Sales']]
unique_region = sps_data_regionVsCat.Region.unique()
uniCat = sps_data_regionVsCat.Category.unique()
colVal = []
for region in unique_region:
    for cat in uniCat:
        total_region_sales = sps_data_regionVsCat.loc[(sps_data_regionVsCat['Category'] == cat) & ( sps_data_regionVsCat['Region'] == region)]['Sales'].sum()
        colVal.append(total_region_sales)
    d[region] = colVal
    colVal = []

d
```

```python
d.to_csv('Prepared_Category_sales')
```
